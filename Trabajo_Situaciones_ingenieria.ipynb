## **Ejercicio 1** "Capacidad de Silos de Granos"
Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico. El usuario ingresa el radio y la altura del silo en metros.


*   *Datos:*  Densidad del maíz ≈0.75 ton/m3
. Volumen V=π×r2×h
*  *Prueba:* Radio = 3m, Altura = 10m. Salida ≈212.06
 toneladas.







```python
import math # Importamos la librería para usar el valor preciso de PI

# ENTRADA DE DATOS
radio = float(input("Ingrese el radio del silo (m): ")) # Entrada str convertida a float
altura = float(input("Ingrese la altura del silo (m): ")) # Entrada str convertida a float

# PROCESO (CÁLCULOS ARITMÉTICOS)
volumen = math.pi * (radio ** 2) * altura # Uso de operadores: potencia (**) y mult (*) (float)
capacidad = volumen * 0.75 # Operador de multiplicación para magnitud física (float)

# SALIDA DE RESULTADOS
print("La capacidad total es:", capacidad, "toneladas") # Salida combinada (str y float)
```

En este ejercicio, el reto fue transformar dimensiones espaciales en unidades de inventario.

*La Lógica:* Primero calculamos cuánto espacio ocupa el grano (Volumen). Usamos



la constante math.pi para que el cálculo sea exacto. El operador ** 2 es fundamental porque el área de la base crece al cuadrado del radio.

*La Transformación: *Una vez tenemos los m
3
 , usamos la Densidad (ρ). Multiplicar el volumen por 0.75 es lo que nos permite pasar de una medida de "espacio" a una medida de "masa" (toneladas).

Dato Clave: Si  se hubiera usado int() en lugar de float(), el programa ignoraría los decimales y en un silo de gran tamaño, eso significaría perder toneladas de registro.

# **Ejercicio 2** " Dosificación de Fertilizante Líquido"
Un sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea. El usuario ingresa las hectáreas totales y la dosis requerida (litros/hectárea).


*   *Datos* :El fertilizante se diluye en agua en una relación 1:500 (1 litro de producto por 500 de agua).
*  *Prueba*: 5 hectáreas, Dosis 2 L/ha. Salida: 10L de producto y 5000L de agua total.




```python
# ENTRADA DE DATOS
hectareas = float(input("Ingrese las hectáreas totales: ")) # str a float
dosis_ha = float(input("Ingrese la dosis (L/ha): ")) # str a float

# PROCESO (CÁLCULOS DE MAGNITUDES)
fertilizante_puro = hectareas * dosis_ha # Operador (*) genera un float
agua_total = fertilizante_puro * 500 # Multiplicación por factor de dilución (float)

# SALIDA DE RESULTADOS
print("Cantidad de producto necesario:", fertilizante_puro, "L") # Salida float
print("Cantidad de agua para dilución:", agua_total, "L") # Salida float
```

*La Lógica*: El programa primero calcula el requerimiento neto de producto comercial. Es un escalamiento lineal: a más hectáreas, más producto.

*La Transformación*: El dato 1:500 es una constante de proporcionalidad. Al multiplicar el fertilizante por 500, el código simula lo que pasaría en el tanque de mezcla: por cada litro que agregas, el sistema debe añadir 500 de agua para que la concentración sea la adecuada y no quemar el cultivo por exceso de sales.

*Dato Clave:* En Python, cuando multiplicas un float (dosis) por un int (500), el resultado es automáticamente un float, manteniendo la precisión de los mililitros.

# **Ejercicio 3** "Rendimiento de Pulpa en Frutas"
Una planta procesadora necesita saber el porcentaje de aprovechamiento. Se ingresa el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas).


*   *Datos*: Peso Total−Residuos
. %Rendimiento=(Pulpa/Peso Total)×100

*   *Peso Total* = 500kg, Residuos = 150kg. Salida: 350kg de pulpa (70%).







```python
# ENTRADA DE DATOS
peso_total = float(input("Ingrese el peso total de la fruta (kg): ")) # str a float
peso_residuos = float(input("Ingrese el peso de los residuos (kg): ")) # str a float

# PROCESO (OPERADORES ARITMÉTICOS)
pulpa = peso_total - peso_residuos # Operador (-) resta de magnitudes
rendimiento = (pulpa / peso_total) * 100 # Operadores (/) división y (*) multiplicación

# SALIDA DE RESULTADOS
print("La pulpa obtenida es:", pulpa, "kg") # Salida float
print("El rendimiento del proceso es:", rendimiento, "%") # Salida float
```

*La Lógica:* Es una operación de resta simple (-) para encontrar la variable de interés (la pulpa). En agroindustria, esto separa el "valor agregado" del "residuo".

*La Transformación*: El cálculo del porcentaje es una normalización. Al dividir pulpa / peso_total, obtenemos un número entre 0 y 1 (tanto por uno). Al multiplicar por 100, lo llevamos a una escala porcentual, que es el lenguaje que entiende el gerente de producción para medir la eficiencia.

*Dato Clave*: El uso de paréntesis en (pulpa / peso_total) * 100 es vital. Por la jerarquía de operaciones en Python, primero queremos asegurar la división antes de escalar el número a porcentaje.

# **Ejercicio 5** "Peso de una Viga de Acero"
Un ingeniero necesita calcular el peso propio de una viga de sección rectangular para el análisis estructural. Se ingresa base, altura y longitud en metros.


*   *Datos*: Densidad del acero =7850 kg/m3
. Peso=Base×Altura×Longitud×Densidad

*  *Prueba*: Base = 0.3m, Altura = 0.5m, Largo = 6m. Salida: 7065 kg.




```python
# ENTRADA DE DATOS (MAGNITUDES FÍSICAS)
base = float(input("Ingrese la base de la viga (m): ").replace(',', '.')) # str a float
altura = float(input("Ingrese la altura de la viga (m): ").replace(',', '.')) # str a float
largo = float(input("Ingrese el largo de la viga (m): ").replace(',', '.')) # str a float

# PROCESO (CÁLCULO ESTRUCTURAL)
volumen = base * altura * largo # Operador (*) para obtener m3
peso_viga = volumen * 7850 # Multiplicación por densidad del acero

# SALIDA DE RESULTADOS
print("El peso propio de la viga es:", peso_viga, "kg") # Salida float
```

*La Lógica*: Este es un cálculo de Carga Muerta. En ingeniería, antes de ponerle peso a una estructura (como sacos de arroz o maquinaria), debemos saber cuánto pesa la estructura misma. La fórmula es una extensión del concepto de Densidad ($\rho = m/V$); simplemente despejamos la masa: $m = V \times \rho$.

*La Transformación*: El programa realiza una integración geométrica. Al multiplicar las tres dimensiones (base * altura * largo), el código construye un "modelo virtual" del volumen del acero. Al aplicar el factor 7850, le "asigna" una realidad física basada en las propiedades del material.

*Dato Clave*: El orden de los factores no altera el producto en la multiplicación (*), pero en programación, agrupar las variables de entrada al principio permite que el código sea más limpio y fácil de auditar por otro ingeniero que deba revisar tus cálculos de carga.

# **Ejercicio 6** "Pendiente de una Rampa de Acceso"
 Determinar si una rampa cumple con la normativa de accesibilidad. El usuario ingresa la altura a vencer y la longitud horizontal de la rampa (en metros).


*   *Datos*: %Pendiente=(Altura/Longitud)×100

*   *Prueba*: Altura = 0.6m, Longitud = 5m. Salida: 12% de pendiente.





```python
# ENTRADA DE DATOS (GEOMETRÍA DE LA RAMPA)
altura = float(input("Ingrese la altura a vencer (m): ").replace(',', '.')) # str a float
longitud = float(input("Ingrese la longitud horizontal (m): ").replace(',', '.')) # str a float

# PROCESO (CÁLCULO DE PENDIENTE)
# La división genera la razón y la multiplicación escala a porcentaje
pendiente = (altura / longitud) * 100 # Uso de operadores (/) y (*)

# SALIDA DE RESULTADOS
print("La rampa tiene una pendiente del:", pendiente, "%") # Salida float
```

*La Lógica*: Este cálculo es una aplicación directa de la tangente del ángulo de inclinación. En ingeniería agroindustrial, este dato nos dice si un operario puede subir una carretilla de forma segura. Si la rampa supera el $10\%$ o $12\%$ (dependiendo de la norma), se considera peligrosa o no apta para acceso universal.

*La Transformación*: El programa realiza una normalización de dimensiones. Al dividir la altura por la longitud, eliminamos las unidades de metros y obtenemos una "razón" pura. Al multiplicar por $100$, llevamos esa razón a una escala porcentual, que es el "idioma" oficial de los planos arquitectónicos.

*Dato Clave*: Al igual que en el ejercicio del rendimiento de pulpa, el uso de paréntesis en (altura / longitud) * 100 es fundamental para la jerarquía de operaciones. Aunque en este caso específico (división y multiplicación) el orden no alteraría el resultado, en fórmulas de ingeniería más complejas, no usar paréntesis podría causar errores catastróficos en el diseño.
